```python
# prompt: translate the image with matrix

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/mm.jpg')

# Get the image dimensions
(h, w) = image.shape[:2]

# Define the translation matrix
tx = 50  # Horizontal translation amount (pixels)
ty = 30  # Vertical translation amount (pixels)
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image
translated_image = cv2.warpAffine(image, M, (w, h))

# Display the translated image
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.show()
```

```python
# prompt: cv code for flip a image vertically and horizontally and both ways and show two two by images with scale values  and remove comment lines

import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/mm.jpg')

# Vertical flip
flipped_vertical = cv2.flip(image, 0)

# Horizontal flip
flipped_horizontal = cv2.flip(image, 1)

# Both vertical and horizontal flip
flipped_both = cv2.flip(image, -1)

# Display the original and flipped images
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB))
plt.title('Vertically Flipped Image')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB))
plt.title('Horizontally Flipped Image')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(flipped_both, cv2.COLOR_BGR2RGB))
plt.title('Both Flipped Image')

plt.show()

# Scale the image (e.g., scale down by 0.5)
scale_percent = 50  # percent of original size
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)

resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

# Display the original and resized images with scale values
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title(f'Resized Image (Scale: {scale_percent}%)')

plt.show()

```

```python
# prompt: perspective image

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('/mm.jpg')

# Define the source points (corners of the original rectangle)
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])

# Define the destination points (corners of the desired rectangle)
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

# Get the perspective transformation matrix
matrix = cv2.getPerspectiveTransform(pts1,pts2)

# Apply the perspective transformation
result = cv2.warpPerspective(image,matrix,(300,300))

# Display the original and perspective transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Perspective Transformed Image')

plt.show()
```

```python
# prompt: cv code for original,translated,scaled,riotated,sheaved

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/mm.jpg')

# Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 5, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Translated Image
(h, w) = image.shape[:2]
tx = 50  # Horizontal translation amount (pixels)
ty = 30  # Vertical translation amount (pixels)
M = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(image, M, (w, h))
plt.subplot(1, 5, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')


# Scaled Image
scale_percent = 50  # percent of original size
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt.subplot(1, 5, 3)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Scaled Image')


# Rotated Image
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
angle = 40
scale = 1.0
M = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(image, M, (w, h))
plt.subplot(1, 5, 4)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')


# Sheared Image
rows, cols = image.shape[:2]
M = np.float32([[1, 0.5, 0],
              [0, 1  , 0],
              [0, 0  , 1]])
sheared_image = cv2.warpPerspective(image,M,(int(cols*1.5),int(rows*1.5)))
plt.subplot(1, 5, 5)
plt.imshow(cv2.cvtColor(sheared_image, cv2.COLOR_BGR2RGB))
plt.title('Sheared Image')


plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload the video file
uploaded = files.upload()
filename = next(iter(uploaded))

# Open the video file
video = cv2.VideoCapture(filename)

# Check if the video file opened successfully
if not video.isOpened():
    print("Error opening video file.")
else:
    # Get video properties
    fps = video.get(cv2.CAP_PROP_FPS)
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Frames per second: {fps}")
    print(f"Width: {width} pixels")
    print(f"Height: {height} pixels")
    print(f"Total frames: {total_frames}")

    # Display frames
    while True:
        ret, frame = video.read()

        if not ret:
            break  # End of video

        # Display the frame (using matplotlib for better inline display in Colab)
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.show()

        # Optionally: Add a delay or a break condition
        # Example: break after displaying 10 frames
        # if video.get(cv2.CAP_PROP_POS_FRAMES) >= 10:
        #     break

    # Release the video capture object
    video.release()
```

```python
# prompt:  write a code to show the laplacian image

import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Load the image in grayscale
image = cv2.imread('/mm.jpg', 0)  # 0 for grayscale

# Calculate the Laplacian
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# Convert the Laplacian image back to 8-bit for display
laplacian = np.uint8(np.absolute(laplacian))

# Display the Laplacian image using matplotlib
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Image')
plt.show()
```
